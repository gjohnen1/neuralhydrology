# Preprocess Historical Data
This notebook converts the basin-specific CSV files into NetCDF files compatible with `GenericDataset`.

```python
import pandas as pd
import xarray as xr
from pathlib import Path
import numpy as np
from tqdm import tqdm

# Configuration
DATA_DIR = Path("../data/harz")
TIMESERIES_DIR = DATA_DIR / "timeseries"
OUTPUT_DIR = TIMESERIES_DIR 

# Ensure output directory exists
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
```

```python
# Find all basin CSVs
csv_files = list(TIMESERIES_DIR.glob("hydromet_timeseries_*.csv"))
print(f"Found {len(csv_files)} CSV files.")

for csv_file in tqdm(csv_files):
    basin_name = csv_file.stem.replace("hydromet_timeseries_", "")
    
    # Read CSV
    df = pd.read_csv(csv_file)
    
    # Ensure date column is datetime
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'])
        df = df.set_index('date')
    else:
        print(f"Warning: No 'date' column in {csv_file}, skipping.")
        continue
        
    # Convert to xarray
    ds = xr.Dataset.from_dataframe(df)
    
    # Save to NetCDF
    output_path = OUTPUT_DIR / f"{basin_name}.nc"
    ds.to_netcdf(output_path)
    print(f"Saved {output_path}")
```
